```python
file_path = "/Users/vinodhinishriram/stocks/nifty monthly Scanlog (2).csv"
headers = ["date_time","no_of_stocks","stocks"]
```

```python
def find_stock_list_pos(str_line):
    n = 0
    p = 0
    for i in range(2):
        p = str_line.find(",")
        n += p
        str_line = str_line[p+1:]
    return n + 1
```

```python
fp = open(file_path,'r')
l = []
i =0 
for line in fp.readlines():
    line = line.strip()
    pos = find_stock_list_pos(line)
    rep_string = line[pos+1:].replace(",",";")
    line = line[:pos+1] + rep_string
    #print(line)
    l.append(line.split(","))
    
```

```python
df = pd.DataFrame(l, columns = headers)
df.head()
```

```python
def count_pe_ce(row, count_string):
    stocks = row['stocks'].split(";")
    count = 0
    for i in stocks:
        if i[-2:] == count_string:
            count += 1
    return count
```

```python
df['pe_count'] = df.apply(lambda row: count_pe_ce(row, 'PE'), axis = 1)
df['ce_count'] = df.apply(lambda row: count_pe_ce(row, 'CE'), axis = 1)
```
